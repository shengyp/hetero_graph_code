```python
from sklearn.manifold import TSNE
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
fig = plt.figure(figsize=(20,20))

plt.rcParams['savefig.dpi'] = 300 #图片像素
plt.rcParams['figure.dpi'] = 300 #分辨率

tsne=TSNE()
```

# =====================================================

```python
cgatdata = np.load('test1.npy')
cgatdata = np.squeeze(cgatdata)

lable_onehot = np.load('one_hot_labels.npy').astype(int)
label = lable_onehot.argmax(axis=1)

tsne.fit_transform(cgatdata)

tsne=pd.DataFrame(tsne.embedding_,index=[i for i in range(cgatdata.shape[0])]) 

tsne['聚类类别'] = label

d=tsne[tsne[u'聚类类别']==0]
plt.plot(d[0],d[1],'r.',markersize=2)

d=tsne[tsne[u'聚类类别']==1]
plt.plot(d[0],d[1],'g.',markersize=2)

d=tsne[tsne[u'聚类类别']==2]
plt.plot(d[0],d[1],'b.',markersize=2)

d=tsne[tsne[u'聚类类别']==3]
plt.plot(d[0],d[1],'y.',markersize=2)

plt.savefig('31.jpg')

plt.show()
```
